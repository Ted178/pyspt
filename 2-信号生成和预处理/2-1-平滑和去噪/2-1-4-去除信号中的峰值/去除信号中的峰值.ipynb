去除信号中的峰值

```python
import numpy as np
from matplotlib import pyplot as plt
import scipy.signal as signal
```

```python
def sinwave(A, f, fs,t):
    Ts = 1 / fs
    n = t / Ts
    n = np.arange(n)
    y = A * np.sin(2 * np.pi * f * n *Ts)
    return y
```

```python
A = 0.5
f = 60
fs =1000
t = 2
T = np.arange(0, 2, 0.001)
```

```python
hz_60 = sinwave(A, f, fs, t)
```

```python
plt.plot(T, hz_60)
plt.axis([-0.05, 2.05, -1, 1])
plt.show()
```

```python
spikeSignal = np.zeros(len(hz_60))
spks = np.arange(10, 1990, 100)
randadd = np.random.randint(-10, 10, size=len(spks))
spikeSignal[spks + randadd] = np.random.randn(len(spks))
```

```python
noisySignal = hz_60 + spikeSignal
```

```python
plt.plot(T, noisySignal)
plt.xlabel('Time (s)')
plt.ylabel('Voltage (V)')
plt.title('Voltage with Added Spikes')
plt.show()
```

```python
medfiltSignal = signal.medfilt(noisySignal, kernel_size=3)
```

```python
plt.plot(T, medfiltSignal)
plt.axis([-0.05, 2.05, -1, 1])
plt.xlabel('Time (s)')
plt.ylabel('Voltage (V)')
plt.title('Voltage with Added Spikes')
plt.grid()
plt.show()
```
